```python
URL = 'https://tiki.vn/'

response = requests.get(URL)
soup = BeautifulSoup(response.text)
```

```python
categories = soup.find_all('a', {'class','efuIbv'})
cat_links = []
for i in range(len(categories)):
    cat_links.append(categories[i]['href'])
```

```python
for link in cat_links[:2]:
    r = requests.get(link).text
    s = BeautifulSoup(r)
    articles = s.find_all('div', {"class" : "product-item"})
    for k in range(len(articles)):
        try:
            #append into arrays
            images = articles[k].img['src']
            fprice = articles[k].find_all("span",{"class":"final-price"})[0].text.strip().split()[0]
            rprice = articles[k].find_all("span",{"class":"price-regular"})[0].text
            discount = ['None' if len(articles[k].find_all("span",{"class":"final-price"})[0].text.strip().split()) == 1 else articles[k].find_all("span",{"class":"final-price"})[0].text.strip().split()[1]][0]
            seller = articles[k]['data-brand']
            titles = articles[k].a['title'].strip().replace('\'','')
            category = articles[k]['data-category'].strip()
            num_reviews = [articles[k].find_all('p',{"class":'review'})[0].text.strip('\(\)') if articles[k].find_all('p',{"class":'review'}) != [] else 'Chưa có nhận xét'][0]
            ratings = [articles[k].find_all('span',{"class":'rating-content'})[0].find('span')['style'].split(':')[1] if articles[k].find_all('span',{"class":'rating-content'}) != [] else 'Rating not available'][0]
            tikinow = ['NO' if articles[k].find_all('i',{"class":"tikicon icon-tikinow-20"}) == [] else 'YES'][0]
            productlink = articles[k].a['href']
            query = f"""INSERT INTO products(images, fprice, category, titles, seller, rprice, discount, ratings, num_reviews, tikinow, productlink)
                        VALUES('{images}','{fprice}','{category}','{titles}','{seller}','{rprice}','{discount}','{ratings}','{num_reviews}','{tikinow}','{productlink}');"""
            cur.execute(query)
            conn.commit()
    #         print(query)
        except Exception as err:
            
            print(err)
```

```python
class product:
    def __init__(self, productid, images,fprice, category, titles, seller, rprice, discount, ratings, num_reviews, tikinow, productlink):
        self.productid  = productid
        self.images = images
        self.fprice = fprice
        self.category = category
        self.titles = titles
        self.seller = seller
        self.rprice = rprice
        self.discount = discount
        self.ratings = ratings
        self.num_reviews = num_reviews
        self.tikinow = tikinow
        self.productlink = productlink
    def insert_db(self):
        query = f"""
            INSERT INTO categories (name, url, parent_id) 
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s) RETURNING id;
        """
        val = (self.productid, self.images, 
               self.fprice,self.category, 
               self.titles, self.seller,
               self.rprice,self.discount,
               self.ratings,self.num_reviews, 
               self.tikinow, self.productlink)
        try:
            cur.execute(query, val)
            # Get id of the new row
            self.cat_id = cur.fetchone()[0]
        except Exception as err:
            print(f'ERROR: {err}')
        
    def __repr__(self):
        return f'ProductID: {self.productid}, Category: {self.category}, Name: {self.titles}, Price: {self.rprice}'
```
